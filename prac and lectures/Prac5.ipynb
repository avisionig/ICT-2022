```python
import requests
```

```python
headers = {'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:105.0) Gecko/20100101 Firefox/105.0'}
resp = requests.get('https://finance.yahoo.com/', headers = headers)
print(resp)
```

```python
headers = {'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:105.0) Gecko/20100101 Firefox/105.0'}
resp = requests.get("https://www.google.com", headers = headers)
print(resp.content)
```

```python
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'}
resp = requests.get('https://query1.finance.yahoo.com/v7/finance/download/TSLA?period1=1633562627&period2=1665098627&interval=1d&events=history&includeAdjustedClose=true',
                    headers = headers)
print(resp.content)
```

```python
import pandas as pd
import io
df = pd.read_csv(io.StringIO(resp.content.decode('utf-8')))
```
