```python
import pandas as pd
import numpy as np
```

```python
a = ['ERROR', '-']
df = pd.read_csv("Downloads/Apple_price_to_clean.csv", na_values = a)
print(df)
```

```python
import datetime
for index, lines in df.iterrows():
    try:
        a = datetime.datetime.strptime(lines['Date'], '%Y-%m-%d')
    except:
        print(lines['Date'])
        df.loc[index, 'Date'] = np.nan
print(df)
```

```python
df.isnull().sum().sum()
```

```python
df.fillna(method = 'bfill', inplace = True)
print(df)
```

```python
df.drop_duplicates(inplace = True)
print(df)
```

```python
df.plot(y = 'High')
df.plot(y = 'Low')
df.plot(y = 'Close')
df.plot(y = 'Volume')
```

```python
print(df.describe())
```

```python
df = df[df['High'] > 20]
df = df[df['Low'] < 1400]
df = df[df['Close'] > 0]
df = df[df['Volume'] < 1.0e+10]
df.plot(y = 'High')
df.plot(y = 'Low')
df.plot(y = 'Close')
df.plot(y = 'Volume')
```
