```python
import numpy as np
import sklearn.linear_model as skmod
import sklearn.preprocessing as skprepro
import sklearn.model_selection as sksel
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = pd.read_parquet(r"Downloads\Data_to_analyse.parquet")
```

```python
print(file)
```

```python
x1 = file["Age"].to_numpy().reshape(-1, 1)
x2 = file["Account_money"].to_numpy().reshape(-1, 1)
y1 = file["Money_given"].to_numpy().reshape(-1, 1)
x = np.hstack([x1, x2])
```

```python
print(x)
```

```python
data_x_train, data_x_test, data_y_train, data_y_test =sksel.train_test_split(x, y1, train_size = 0.8,shuffle = True)

# plt.scatter(data_x_train, data_y_train)
# plt.scatter(data_x_test, data_y_test)
```

```python
stanScaler = skprepro.StandardScaler()
```

```python
stan1 = stanScaler.fit_transform(data_x_train)
labelScaler = skprepro.StandardScaler()
label_stan = labelScaler.fit_transform(data_y_train)
print(stan1)
```

```python
print(label_stan)
```

```python
model = skmod.LinearRegression().fit(stan1, label_stan)
```

```python
print(model.coef_)
print(model.intercept_)
print(model.score(stan1, label_stan))
#y' = (-0.860)*x1  +  0.353 * x2 + (-2.57400011e-17)
```

```python
stanScaler = skprepro.StandardScaler()
```

```python
stan1 = stanScaler.fit_transform(data_x_test)
labelScaler = skprepro.StandardScaler()
label_stan = labelScaler.fit_transform(data_y_test)
```

```python
model = skmod.LinearRegression().fit(stan1, label_stan)
```

```python
print(model.coef_)
print(model.intercept_)
print(model.score(stan1, label_stan))
# y' = (-0.836) * x1 + 0.327 * x2 + (-4.320)
```
