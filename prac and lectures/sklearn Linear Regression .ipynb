```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.linear_model as skmod
```

```python
x = [6, 10, 2, 3, 4, 0, 7, 8, 9, 1]
y = [130, 21, 43, 76, 105, 3, 167, 162, 91, 15]
```

```python
x1 = []
x2 = []
for i in x:
    x1.append(i * i)
    x2.append(i * i * i)
print(x1)
print(x2)
```

```python
x = np.array(x).reshape(10,1)
print(x)
y = np.array(y).reshape(10,1)
print(y)
x1 = np.array(x1).reshape(10,1)
print(x1)
x2 = np.array(x2).reshape(10,1)
print(x2)
```

```python
model = skmod.LinearRegression()
```

```python
model_x = model.fit(x,y)
```

```python
print(model_x.coef_)
print(model_x.intercept_)
print(model_x.score(x,y))
```

```python
plt.scatter(x, y)
plt.plot([0,10], [8.709*0 + 37.754, 8.709 *10 + 37.754])
#y' = 8.709 * x + 37.754
```

```python
xqs = np.hstack([x,x1])
xcube = np.hstack([x,x1,x2])
print(xqs)
print(xcube)
```

```python
model = skmod.LinearRegression()
model_sq = model.fit(xqs, y)
```

```python
print(model_sq.coef_)
print(model_sq.intercept_)
print(model_sq.score(xqs,y))
# y' = 56.77*x1 - 4.80 *x1^2 - 29.53
```

```python
model_cube = model.fit(xcube, y)
```

```python
print(model_cube.coef_)
print(model_cube.intercept_)
print(model_cube.score(xcube,y))
#y' = 1.47 * x1 + 9.69 * x2 - 0.96 * x3 + 5.27
```

```python
plt.scatter(x, y)
x_pred = np.arange(0, 10, 0.1)
x_pred = x_pred.reshape(-1,1)
x_pred2 = x_pred * x_pred 
x_p_f = np.hstack([x_pred, x_pred2])
y_pred = model_sq.predict(x_p_f)
plt.plot(x_pred, y_pred)

plt.plot([0,10], [8.709*0 + 37.754, 8.709 *10 + 37.754])
```
